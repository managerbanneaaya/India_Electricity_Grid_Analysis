# 1. IN-WE Zone 2024


### Electricity Mix Hourly (MW)

```python
import requests
import pandas as pd
from datetime import datetime, timedelta
from getpass import getpass
API_KEY = getpass("API_KEY")


ZONE = "IN-WE"  # change if needed
BASE_URL = "https://api.electricitymaps.com/v3/electricity-mix/past-range" 

start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 12, 31)

step = timedelta(days=10)

all_data = []

current_start = start_date

while current_start < end_date:
    
    current_end = min(current_start + step, end_date)
    
    start_str = current_start.strftime("%Y-%m-%dT%H:%MZ")
    end_str = current_end.strftime("%Y-%m-%dT%H:%MZ")
    
    params = {
        "zone": ZONE,
        "start": start_str,
        "end": end_str
    }
    
    headers = {
        "auth-token": API_KEY
    }
    
    print(f"Fetching {start_str} -> {end_str}")
    
    response = requests.get(BASE_URL, headers=headers, params=params)
    
    if response.status_code == 200:
        data = response.json()
        all_data.extend(data["data"])
    else:
        print("Error:", response.status_code, response.text)
    
    current_start += step

# Convert to DataFrame
df = pd.json_normalize(all_data)

# Save dataset
df.to_csv("electricity_mix_2024_hourly.csv", index=False)

print("Finished. Total rows:", len(df))
```
